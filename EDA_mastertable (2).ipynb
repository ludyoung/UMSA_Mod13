### DistribuciÃ³n de la satisfacciÃ³n del cliente

### Matriz de correlaciones

```python
num_cols = df.select_dtypes(include=["number"]).columns

corr_matrix = df[num_cols].corr()

# CorrelaciÃ³n de cada variable con cust_review_mean
target_corr = corr_matrix["cust_review_mean"].sort_values(ascending=False)

print("\nðŸ”¹ Correlaciones con cust_review_mean (ordenadas):")
display(target_corr)

# Opcional: mostrar heatmap de las N variables mÃ¡s correlacionadas
top_n = 15  # puedes cambiar este valor
top_features = target_corr.abs().sort_values(ascending=False).head(top_n).index

plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix.loc[top_features, top_features],
            annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Matriz de correlaciÃ³n (top variables mÃ¡s relacionadas con cust_review_mean)")
plt.show()
```

### DistribuciÃ³n de calificaciones por fecha

```python
import pandas as pd

# Asegurar tipo datetime
df["last_review_date"] = pd.to_datetime(df["last_review_date"], errors="coerce")

# CalificaciÃ³n entera (1â€“5) para usar en grÃ¡ficos
df["cust_review_int"] = df["cust_review_mean"].round().astype("Int64")

# Columnas de aÃ±o y mes
df["review_year"] = df["last_review_date"].dt.year
df["review_month"] = df["last_review_date"].dt.month

# Columna aÃ±o-mes para agregados mensuales
df["year_month"] = df["last_review_date"].dt.to_period("M")
```

```python
import matplotlib.pyplot as plt

# Agrupar por aÃ±o-mes y calificaciÃ³n entera
monthly_counts = (
    df
    .groupby(["year_month", "cust_review_int"])
    .size()
    .reset_index(name="count")
)

# Pasar a formato tabla para grÃ¡fico apilado
pivot_monthly = (
    monthly_counts
    .pivot(index="year_month", columns="cust_review_int", values="count")
    .fillna(0)
    .sort_index()
)

# Ordenar columnas por calificaciÃ³n (1â€“5)
pivot_monthly = pivot_monthly.reindex(sorted(pivot_monthly.columns), axis=1)

# Graficar
plt.figure(figsize=(14, 5))
pivot_monthly.plot(kind="bar", stacked=True, ax=plt.gca())
plt.title("DistribuciÃ³n de calificaciones por mes")
plt.xlabel("AÃ±o-mes")
plt.ylabel("Cantidad de reviews")
plt.tight_layout()
plt.show()
```
